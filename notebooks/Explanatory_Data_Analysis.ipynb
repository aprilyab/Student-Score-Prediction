#  Explanatory Data Analysis (EDA) — Student Performance Dataset

**Objective:**  
Analyze the *Student Performance Factors* dataset and understand the variables affecting student exam performance.  

We will:  
1. Load the dataset  
2. Explore basic information (shape, types, missing values)  
3. Perform descriptive statistics  
4. Visualize target variable (Exam_Score)  
5. Explore relationships between study hours and scores  
6. Check correlations among features  


## Import packages and Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set(style="whitegrid", palette="muted")
```

## Load the Raw Dataset

```python
df=pd.read_csv("../data/raw/StudentPerformanceFactors.csv")
```

```python
# Information about columns and datatypes
print(df.info())

# Summary statistics
print(df.describe().T)


# Check for missing values
print(df.isnull().sum())
```

```python
df
```

```python
# Distribution of Final Scores (Target: G3)
plt.figure(figsize=(8,6))
sns.histplot(df["Exam_Score"],bins=20,kde=True)
plt.title("the distribution of final exam score")
plt.ylabel("count")
plt.xlabel("exam score")
plt.show()
```

```python
# Study Time vs Final Score
plt.figure(figsize=(8,6))
sns.scatterplot(x="Hours_Studied",y="Exam_Score",data=df)
plt.xlabel("study time")
plt.ylabel("exam score")
plt.show()
```

```python
# Boxplot: Study Time Categories
plt.figure(figsize=(8,6))
sns.boxplot(x="Hours_Studied",y="Exam_Score",data=df)
plt.title("Boxplot of Study Time vs Final Score")
plt.xlabel("Study Time Category")
plt.ylabel("Final exam score")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
